```python
import pandas as pd

# טעינת הנתונים
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")  # עדכן אם השם שונה

# הצגת מספר שורות ועמודות
print(f"מספר השורות: {df.shape[0]}")
print(f"מספר העמודות: {df.shape[1]}")
```

```python
print(df.dtypes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def reverse_text(text):
    return text[::-1]

sns.histplot(df["Sleep Duration"], bins=20, kde=True)
plt.title(reverse_text("התפלגות משך השינה"))
plt.xlabel(reverse_text("שעות שינה"))
plt.ylabel(reverse_text("מספר אנשים"))
plt.show()
```

In this section, I created a histogram to show the distribution of sleep duration among people in the dataset. I used the Seaborn and Matplotlib libraries to build the char

```python
sns.scatterplot(x="Physical Activity Level", y="Sleep Duration", data=df)
plt.title(reverse_text("קשר בין פעילות גופנית למשך השינה"))
plt.xlabel(reverse_text("רמת פעילות גופנית"))
plt.ylabel(reverse_text("משך שינה"))
plt.show()
```

I created a scatter plot to explore the relationship between physical activity level and sleep duration

```python
import seaborn as sns
import matplotlib.pyplot as plt

def reverse_text(text):
    return text[::-1]

# שינוי גודל הגרף
plt.figure(figsize=(10, 6))

# התאמת צבעים – ניתן לבחור סטייל מתוך seaborn
sns.set_style("whitegrid")

# יצירת גרף ויולין עם צבעים מותאמים
sns.violinplot(x="BMI Category", y="Heart Rate", data=df, hue="BMI Category", palette="coolwarm", legend=False)

# התאמת כותרות וצירים
plt.title(reverse_text("השוואת דופק לפי קטגוריית BMI"), fontsize=16, fontweight="bold", color="darkblue")
plt.xlabel(reverse_text("קטגוריית BMI"), fontsize=14, color="darkred")
plt.ylabel(reverse_text("דופק"), fontsize=14, color="darkgreen")

# התאמת סיבוב כיתובים בצירים
plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

I created a violin plot to compare heart rate across different BMI categories. I used Seaborn and customized the style, colors, and figure size to make the graph more visually clear.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def reverse_text(text):
    return text[::-1]

plt.figure(figsize=(10, 6))  # הגדלת גודל הגרף

# יצירת קו חלק יותר עם נקודות נתונים להדגשה
sns.lineplot(x="Age", y="Sleep Duration", data=df, linestyle="dashed", marker="o", color="darkblue")

# התאמת כותרות וצירים
plt.title(reverse_text("מגמות שינה לפי גיל"), fontsize=16, fontweight="bold", color="darkred")
plt.xlabel(reverse_text("גיל"), fontsize=14, color="darkgreen")
plt.ylabel(reverse_text("משך שינה"), fontsize=14, color="darkblue")

# התאמת גודל הכיתובים וסיבוב לסדר ברור יותר
plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
sns.catplot(x="BMI Category", hue="Gender", kind="count", data=df)
plt.title(reverse_text("התפלגות קטגוריות BMI לפי מין"))
plt.xlabel(reverse_text("קטגוריית BMI"))
plt.ylabel(reverse_text("מספר אנשים"))
plt.show()
```

 created a count plot to show the distribution of BMI categories separated by gender. This helps me see how BMI categories differ between males and females. I used Seaborn’s catplot

```python
sns.catplot(x="Stress Level", y="Quality of Sleep", hue="Gender", kind="bar", data=df)
plt.title(reverse_text("השוואת איכות שינה לפי לחץ ומגדר"))
plt.xlabel(reverse_text("רמת לחץ"))
plt.ylabel(reverse_text("איכות שינה"))
plt.show()
```

 I created a bar plot to compare sleep quality based on stress level and gender. I used Seaborn’s catplot with bars to see how stress affects sleep quality differently for males and females

```python
sns.lineplot(x="Age", y="Quality of Sleep", hue="Stress Level", data=df, marker="o")
plt.title(reverse_text("השוואת איכות שינה לפי רמות לחץ לאורך זמן"))
plt.xlabel(reverse_text("גיל"))
plt.ylabel(reverse_text("איכות שינה"))
plt.show()
```

In this part, I created a line plot to compare sleep quality across different ages, separated by stress levels. Using Seaborn, I added markers to show data points clearly

```python
sns.relplot(x="Stress Level", y="Sleep Duration", data=df, kind="scatter", hue="Gender")
plt.title(reverse_text("קשר בין משך שינה לרמות לחץ"))
plt.xlabel(reverse_text("רמת לחץ"))
plt.ylabel(reverse_text("משך שינה"))
plt.show()
```

In this section, I created a scatter plot to explore the relationship between stress level and sleep duration, with points colored by gender. I used Seaborn’s relplot to visualize how stress might affect sleep length for males and females.

```python
sns.relplot(x="Physical Activity Level", y="Quality of Sleep", data=df, kind="line", hue="Gender")
plt.title(reverse_text("השפעת פעילות גופנית על איכות שינה"))
plt.xlabel(reverse_text("רמת פעילות גופנית"))
plt.ylabel(reverse_text("איכות שינה"))
plt.show()
```

```python
sns.countplot(x="BMI Category", hue="BMI Category", data=df, palette="coolwarm", legend=False)
plt.title(reverse_text("ספירת אנשים לפי קטגוריית BMI"))
plt.xlabel(reverse_text("קטגוריית BMI"))
plt.ylabel(reverse_text("מספר אנשים"))
plt.show()
```

```python
sns.pointplot(x="BMI Category", y="Sleep Duration", data=df)
plt.title(reverse_text("השפעת BMI על משך השינה"))
plt.xlabel(reverse_text("קטגוריית BMI"))
plt.ylabel(reverse_text("משך שינה"))
plt.show()
```
